```python
import pandas as pd

# Step 1: Create Sample Dataset
Msc026= {
    'Region': ['EAST', 'WEST', 'NORTH', 'SOUTH', 'EAST', 'WEST', 'NORTH' , 'SOUTH', 'EAST', 'WEST'],
    'Product': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Sales': [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000],
    'Quantity': [10, 15, 20, 25, 30, 35, 40, 50, 60, 70],
}
```

```python
df_Msc026 = pd.DataFrame(Msc026)
print("Sample Dataset Msc026:\n", df_Msc026)
```

```python
# Step 2: Grouping and Aggregation
# Aggregating Sales by Region (Sum Aggregation)
sales_by_region_Msc026 = df_Msc026.groupby('Region')['Sales'].sum()
print("\nTotal Sales by Region Msc026:\n", sales_by_region_Msc026)
```

```python
# Aggregating Sales and Quantity by Product (Mean Aggregation)
mean_by_product_Msc026 = df_Msc026.groupby('Product')[['Sales', 'Quantity']].mean()
print("\nMean Sales and Quantity by Product Msc026:\n", mean_by_product_Msc026)
```

```python
# Aggregating Count of Sales by Region (Count Aggregation)
count_by_region_Msc026 = df_Msc026.groupby('Region')['Sales'].count()
print("\nCount of Sales Records by Region Msc026:\n", count_by_region_Msc026)
```

```python
# Custom Aggregation: Calculate Min and Max Sales by Region
custom_aggregation_Msc026 = df_Msc026.groupby('Region')['Sales'].agg(['min', 'max'])
print("\nCustom Aggregation (Min and Max Sales by Region) Msc026:\n", custom_aggregation_Msc026)
```

```python
# Step 3: Multi-Level Aggregation
# Aggregating Sales by Region and Product
multi_level_agg_Msc026 = df_Msc026.groupby(['Region', 'Product'])['Sales'].sum()
print("\nSales by Region and Product Msc026:\n", multi_level_agg_Msc026)
```

```python
# Step 4: Reset Index for Multi-Level Aggregation
multi_level_agg_reset_Msc026 = multi_level_agg_Msc026.reset_index()
print("\nSales by Region and Product (Reset Index) Msc026:\n", multi_level_agg_reset_Msc026)
```

```python
# Step 1: Extend Dataset with Time and Spatial Data
Msc0262 = {
    'Region': ['EAST', 'WEST', 'NORTH', 'SOUTH', 'EAST', 'WEST', 'NORTH' , 'SOUTH', 'EAST', 'WEST'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Kolkata', 'Jaipur','Mumbai', 'Delhi', 'Chennai', 'Kolkata', 'Jaipur'],
    'Product': ['A', 'B', 'A', 'B', 'C', 'C', 'B', 'A', 'A', 'C'],
    'Sales': [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000],
    'Quantity': [10, 15, 20, 25, 30, 35, 40, 45, 50, 55],
    'Date': pd.to_datetime(['2024-01-01', '2024-01-02', '2024-02-01', '2024-02-03',
                            '2024-03-01', '2024-03-02', '2024-04-01', '2024-04-03', '2024-02-01', '2024-02-03'])
}
```

```python
df_Msc0262 = pd.DataFrame(Msc0262)
print("Extended Dataset Msc026:\n", df_Msc026)
```

```python
# -----------------------------
# Time Aggregation
# -----------------------------

# Step 2: Set Date Column as Index (optional)
df_Msc0262.set_index('Date', inplace=True)
```

```python
# Aggregating Sales by Month
monthly_sales_Msc026 = df_Msc0262.resample('M')['Sales'].sum()
print("\nTotal Sales by Month Msc026:\n", monthly_sales_Msc026)
```

```python
# Aggregating Sales by Quarter
quarterly_sales_Msc026 = df_Msc0262.resample('Q')['Sales'].sum()
print("\nTotal Sales by Quarter Msc026:\n", quarterly_sales_Msc026)
```

```python
# Aggregating Sales by Year
yearly_sales_Msc026 = df_Msc0262.resample('Y')['Sales'].sum()
print("\nTotal Sales by Year Msc026:\n", yearly_sales_Msc026)
```

```python
# Reset Index to Restore Original Structure
df_Msc0262.reset_index(inplace=True)
```

```python
# -----------------------------
# Spatial Aggregation
# -----------------------------

# Step 3: Aggregating Sales by Region
sales_by_region_Msc026 = df_Msc0262.groupby('Region')['Sales'].sum()
print("\nTotal Sales by Region Msc026:\n", sales_by_region_Msc026)
```

```python
# Aggregating Sales by City
sales_by_city_Msc026 = df_Msc0262.groupby('City')['Sales'].sum()
print("\nTotal Sales by City Msc026:\n", sales_by_city_Msc026)
```

```python
# Aggregating Sales by Region and City
sales_by_region_city_Msc026 = df_Msc0262.groupby(['Region', 'City'])['Sales'].sum()
print("\nTotal Sales by Region and City Msc026:\n", sales_by_region_city_Msc026)
```
