# Feature Transformation Techniques in Data Mining

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import FunctionTransformer
```

```python
# Step 1: Create sample data
msc026_data = np.array([1, 2, 3, 4, 5, 6], dtype=float)  # Sample data array
```

```python
# Logarithmic transformation
msc026_transform = FunctionTransformer(func=np.log1p)  # Using sklearn's FunctionTransformer
msc026_log_transformed_data = msc026_transform.fit_transform(msc026_data)
```

```python
# Square transformation
msc026_square_transformed_data = np.square(msc026_data)
```

```python
# Reciprocal transformation
msc026_reciprocal_transformed_data = np.reciprocal(msc026_data)
```

```python
# Trigonometric transformations (custom_
msc026_sin_transformed_data = np.sin(msc026_data)
msc026_cos_transformed_data = np.cos(msc026_data)
msc026_tan_transformed_data = np.tan(msc026_data)
```

```python
# Square root transformation
msc026_sqrt_transformed_data = np.sqrt(msc026_data)
```

```python
# Combine into a DataFrame for comparison
msc026_results = pd.DataFrame({
    'Original': msc026_data,
    'Log (log1p)': msc026_log_transformed_data,
    'Square': msc026_square_transformed_data,
    'Reciprocal': msc026_reciprocal_transformed_data,
    'Sin': msc026_sin_transformed_data,
    'Cos': msc026_cos_transformed_data,
    'Tan': msc026_tan_transformed_data,
    'Square Root': msc026_sqrt_transformed_data
})

print(msc026_results)
```
